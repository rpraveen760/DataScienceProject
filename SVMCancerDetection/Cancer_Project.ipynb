```python
# Dispersion of categorical data
print(data['diagnosis'].value_counts())
data['diagnosis'].value_counts().plot(kind='bar')
```

```python
#-labeling
data['diagnosis'] = data['diagnosis'].map({'M':0, 'B':1})
data.head()
```

```python
# dependant and independant variables
y = data['diagnosis']
X = data.loc[:, data.columns != 'diagnosis']
```

```python
# Splitting train-test data

from sklearn.model_selection import train_test_split

# using the train test split function
X_train, X_test, y_train, y_test = train_test_split(X,y ,
								random_state=104,
								test_size=0.2,
								shuffle=True)
```

```python
X_train 
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train) 
X_test = sc_x.transform(X_test)
  
print (X_train[0:10, :])
```

```python
#model

from sklearn.svm import SVC  
clf = SVC(kernel='linear') 
  
# fitting x samples and y classes 
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
y_pred_train = clf.predict(X_train)

# metrics are used to find accuracy or error
from sklearn import metrics
print()

# using metrics module for accuracy calculation
print("TRAIN ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_train, y_pred_train))
print("TEST ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))

#classification report
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import plot_confusion_matrix
print(classification_report(y_test,y_pred))
cf_matrix = confusion_matrix( y_test,y_pred)

```

```python
#plotting the confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues',fmt="d")
ax.set_title(' Confusion Matrix \n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');
## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['Malignqnt','Benign'])
ax.yaxis.set_ticklabels(['Malignant','Benign'])
## Display the visualization of the Confusion Matrix.
plt.show()
```

```python
#HYPER-PARAMETER TUNING

from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
			'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
			'kernel': ['rbf','linear']}

grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)

# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
# print best parameter after tuning
print(grid.best_params_)

# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python
#AFTER-HYPER_TUNING
from sklearn.svm import SVC  
clf = SVC(C = 10 , gamma = 0.01 , kernel='rbf') 
  
# fitting x samples and y classes 
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
y_pred_train = clf.predict(X_train)

# metrics are used to find accuracy or error
from sklearn import metrics
print()

# using metrics module for accuracy calculation
print("TRAIN ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_train, y_pred_train))
print("TEST ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))

#classification report
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import plot_confusion_matrix
print(classification_report(y_test,y_pred))
cf_matrix = confusion_matrix( y_test,y_pred)


```

```python
#plotting the confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues',fmt="d")
ax.set_title(' Confusion Matrix \n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');
## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['Malignant','Benign'])
ax.yaxis.set_ticklabels(['Malignant','Benign'])
## Display the visualization of the Confusion Matrix.
plt.show()
```

```python
# SENSITIVITY AND SPECIFICITY 
print(cf_matrix)
sensitivity = cf_matrix[0][0] / (cf_matrix[0][0] + cf_matrix[1][0])
specificity = cf_matrix[1][1] / (cf_matrix[1][1] + cf_matrix[0][1])

print('Specificity : ' ,specificity)
print('Senitivity : ' ,sensitivity)

  
```

```python
# AUC - ROC 
from sklearn.metrics import  roc_auc_score, roc_curve

def plot_roc_curve(true_y, y_prob):
    """
    plots the roc curve based of the probabilities
    """

    fpr, tpr, thresholds = roc_curve(true_y, y_prob)
    plt.plot(fpr, tpr)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('AUC')
```

```python
plot_roc_curve(y_test, y_pred)
print(f'SVM-model AUC score: {roc_auc_score(y_test, y_pred)}')
```
